Series: Tabelas de apenas 1 coluna

```python
#Quando for uma tabela de apenas uma coluna, entenderemos ela como sendo uma SERIE (array unidimensional)
#A serie já tem um indice automático 
minha_lista = ["A","B","C","D","E"]
minha_serie = pd.Series(minha_lista)

#Posso filtrar tanto pelo index quanto pelos valores
minha_serie[minha_serie > "B"]
```

```python
#Também é possível determinar um índice customizado
minha_lista = ["A","B","C","D","E"]
meu_index = [5,6,3,8,9]
minha_serie = pd.Series(minha_lista, index=meu_index)
minha_serie
```

IF: Análise Condicional em Dataframe

```python
def funcao_cond(parametro):
    if parametro >=100:
        resultado = 'maior que 100'
    elif (parametro > 80) & (parametro <100):
        resultado = 'maior que 80 e menor que 100'
    else:
        resultado = 'menor que 80'
    return resultado

df_vendas['Resultado'] = df_vendas['Produto_C'].apply(funcao_cond)
df_vendas
```

df.concat: Union/Append de dois dataframes

```python
df_um = [{'pais': 'Brasil','uf':'Rio de Janeiro', 'cliente': 'Rodrigo'}, 
{'pais': 'Brasil', 'uf':'Rio de Janeiro', 'cliente': 'Jonatas'}]


df_dois = [{'pais': 'Brasil','uf':'Sao Paulo',  'cliente': 'Allan'},
{'pais': 'Chile','uf':'Santiago',  'cliente': 'Andy'}]

df_um = pd.DataFrame(df_um)
df_dois = pd.DataFrame(df_dois)

df_union_um_e_dois = pd.concat([df_um,df_dois], axis=0).reset_index(drop=True) #O drop=True faz que seja deletada a coluna de index e criada outra nova
df_union_um_e_dois
```

pd.join: JOIN entre Dataframes

```python
df_A = [{'uf':'Rio de Janeiro', 'cliente': 'Rodrigo'}, #Passei o valor como um dicionário
{'uf':'Rio de Janeiro', 'cliente': 'Jonatas'},
{'uf':'Sao Paulo',  'cliente': 'Allan'},
{'uf':'Santiago',  'cliente': 'Andy'},
{'uf':'Lima',  'cliente': 'Fred'}]

df_B = [{'pais':'Brasil', 'ESTADO':'Rio de Janeiro'},
        {'pais': 'Chile','ESTADO':'Santiago'},
        {'pais': 'Peru','ESTADO':'Lima'}]

df_A = pd.DataFrame(df_A)
df_B = pd.DataFrame(df_B)

#primeiro especifico a chave da tabela esquerda, depois a da tabela direita, usando o set_index 
df_A.join(on='uf', other=df_B.set_index('ESTADO'), how='left') #Outros métodos (inner, right, outer)

```

pd.merge: JOIN entre Dataframes (Melhor opção)

```python
pd.merge(df_A, df_B, left_on='uf',right_on='ESTADO', how='left') #Outros métodos (inner, right, outer)
```
