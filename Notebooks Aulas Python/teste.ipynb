```python
import numpy as np
import matplotlib.pyplot as plt

# Dados de exemplo
categorias = ['Sabor A', 'Sabor B', 'Sabor C', 'Sabor D']
valores = [40, 25, 15, 30]

# N√∫mero de categorias
num_categorias = len(categorias)

# √Çngulos para as categorias
angulos = np.linspace(0, 2 * np.pi, num_categorias, endpoint=False).tolist()
angulos += angulos[:1]  # Fecha o gr√°fico

# Valores para o primeiro e √∫ltimo √¢ngulo para uma conex√£o fechada
valores += valores[:1]

# Crie um gr√°fico de radar
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)

# Ajuste as posi√ß√µes dos r√≥tulos
offset = np.pi / 10  # Ajuste de offset para os r√≥tulos
ax.set_xticks(angulos[:-1])
ax.set_xticklabels(categorias)

# Plote os valores
ax.fill(angulos, valores, 'b', alpha=0.1)

# Ajuste as posi√ß√µes dos r√≥tulos para evitar sobreposi√ß√£o
for label, angle in zip(ax.get_xticklabels(), angulos[:-1]):
    x, y = label.get_position()
    if angle == np.pi / 2:
        x += 0.2  # Afastar r√≥tulo da dire√ß√£o de 90 graus
    elif angle == 3 * np.pi / 2:
        x -= 0.2  # Afastar r√≥tulo da dire√ß√£o de 270 graus
    label.set_position((x, y))

# Adicione um t√≠tulo
plt.title("Gr√°fico de Radar")

# Mostrar o gr√°fico
plt.show()
```
