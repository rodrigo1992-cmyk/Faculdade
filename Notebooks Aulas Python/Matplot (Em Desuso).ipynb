```python
meu_dic = {'Atividades': ['Atividade1','Atividade2','Atividade3'],
       'Dt inicio': ['2023-01-01','2023-01-10','2023-01-20'],
       'Dt fim': ['2023-01-10','2023-01-25','2023-01-30']}

meu_df = pd.DataFrame(meu_dic) #Converter o dicionário para um Dataframe do Pandas

meu_df['Dt inicio'] = pd.to_datetime(meu_df['Dt inicio']) #Converter os campos de data
meu_df['Dt fim'] = pd.to_datetime(meu_df['Dt fim']) #Converter os campos de data

lista_datas = pd.date_range(min(meu_df['Dt inicio']),max(meu_df['Dt fim']))
lista_ticks = list(range(0,len(lista_datas)))

#Começar o Plot do Gantt
plt.barh(meu_df['Atividades'], width=(meu_df['Dt fim'] - meu_df['Dt inicio']).dt.days, left=meu_df['Dt inicio'])
plt.xticks(lista_datas,lista_datas,rotation = 45)
plt.xlabel('Data', rotation = 0)
plt.ylabel('Tarefas')
plt.title('Gráfico de Gantt')
plt.show()
```

Subplots: Grids que contém mais de um gráfico

```python
#Segue abaixo formato da instrução com principais parâmetros
#fig,axes = plt.subplots(nrows, ncols, figsize = (width, height))

#Criando um conjunto de dados qualquer para serem plotados
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-x/5) * np.sin(x)

#Passando a configuração do subplot, que é basicamente o grid
fig,axes = plt.subplots(1,3, figsize =(16,4), sharey = True) #O parâmetro Share Y faz com que aparece o eixo Y somente no primeiro gráfico.

#todos os gráficos que vierem após a instrução do subplot serão entendidos como parte do grid
axes[0].plot(x, y1, label="sin(x)", color="blue")
axes[0].set_title('sin(x)')
axes[0].legend()

axes[1].plot(x, y2, label='cos(x)', color='red')
axes[1].set_title('cos(x)')
axes[1].legend()

axes[2].plot(x, y3, label='exp(-x/5) * np.sin(x)', color='green')
axes[2].set_title('exp(-x/5) * np.sin(x)')
axes[2].legend()


plt.tight_layout() #Só serve para que não haja espaçamento entre os gráficos
plt.show()
```

```python
#Se quiser que o Grid possua linhas e colunas os gráficos tem de ser identificados por uma tupla de[linha,coluna]

fig,axes = plt.subplots(2,2, figsize =(16,4), sharey = True) #O parâmetro Share Y faz com que aparece o eixo Y somente no primeiro gráfico.

#todos os gráficos que vierem após a instrução do subplot serão entendidos como parte do grid
axes[0,0].plot(x, y1, label="sin(x)", color="blue")
axes[0,0].set_title('graf_1')
axes[0,0].legend()

axes[0,1].plot(x, y2, label='cos(x)', color='red')
axes[0,1].set_title('graf_2')
axes[0,1].legend()

axes[1,0].plot(x, y3, label='exp(-x/5) * np.sin(x)', color='green')
axes[1,0].set_title('graf_3')
axes[1,0].legend()

axes[1,1].plot(x, y2, label='cos(x)', color='pink')
axes[1,1].set_title('graf_4')
axes[1,1].legend()

plt.tight_layout() #Só serve para que não haja espaçamento entre os gráficos
plt.show()
```

errorbar: Adicionar margem de erro (Barra de Incerteza)

```python
x = np.array([1,2,3,4,5,6])
y = np.array([10,12,8,9,11,12])
matriz_erro = np.array([0.4, 0.7, 0.3, 0.5, 0.2, 0.6]) #Valores de margem de erro para cada ponto do gráfico

plt.errorbar(x, y, yerr = matriz_erro, fmt='o', color='blue', ecolor ='red', capsize=5, label='Margem de Erro') #Se quiser adicionar a margem de erro no eixo Y basta utilizar o 'xerr' no lugar do 'yerr'

#Eu posso combinar o error bar com quase qualquer gráfico, basta adicionar a instrução do gráfico desejado em seguida, como o abaixo que deixei comentado
#plt.plot(x,y)
plt.legend()
plt.show()
```
