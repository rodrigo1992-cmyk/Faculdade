```python
import kaggle as kag
from kaggle.api.kaggle_api_extended import KaggleApi
```

> Autenticar a API

```python
api = KaggleApi()
api.authenticate()
```

> Realizar uma busca na lista de datasets

```python
api.datasets_list(search ="iris")
```

> Baixar o Dataset desejado

```python
#Basta indicar o final da URL do arquivo (neste caso a url era https://www.kaggle.com/datasets/uciml/iris)
#E apontar um caminho de onde deve ser feito o download do arquivo

api.dataset_download_files("uciml/iris", path="data/")
```

> Realizar o Unzip, pois neste caso o dataset contêm mais de um arquivo

```python
from zipfile import ZipFile
ZipFile("data/iris.zip").extractall("data/")
ZipFile("data/iris.zip").close
```
