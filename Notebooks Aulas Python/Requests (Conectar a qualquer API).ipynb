Como fazer uma requisição

```python
#Ao dar o get irá retornar um objeto com mais de um conteúdo. 
#Se mandar imprimir direto o resultado somenteo código [200] que significa que algum valor foi retornado
conteudo = req.get("https://api.openbrewerydb.org/v1/breweries")
print(conteudo)
```

```python
#Para obter os dados contidos no objeto "conteudo" se faz necessário usar a funçao ".text"
dados = conteudo.text
print(conteudo.text)
```

Como baixar direto em json

```python
#A própria biblioteca request tem função para já baixar direto como json
dados_organizados = req.get("https://api.openbrewerydb.org/v1/breweries").json()
dados_organizados
```

Como passar parâmetros

```python
#Como passar parâmetros
meus_parametros = {
    "by_city":"new_york",
    "per_page": 50
}
dados_filtrados = req.get("https://api.openbrewerydb.org/v1/breweries", params = meus_parametros).json()

dados_filtrados
```

Baixar direto para um arquivo

```python
import json
#O "r" antes do caminho serve para converter uma string comum em um raw path
caminho_arquivo = r'C:\Users\RodrigoPintoMesquita\.vscode\Meus Notebooks\Python\arquivos\meu_arquivo.json'

with open(caminho_arquivo, "w") as meu_arquivo:
    json.dump(dados_organizados, meu_arquivo, indent= True)
```

Analisar os dados

```python
paises_dict = {}
for linha in dados_organizados:
    pais = linha["country"]
    if pais not in paises_dict:
        paises_dict[pais] = 1
    else:
        paises_dict[pais] += 1

paises_dict
```

```python
#Outra forma de contar a quantidade de registro por país

from collections import Counter

lista_paises = []
for linha in dados_organizados:
    lista_paises.append(linha["country"])

Counter(lista_paises)
```
