# Manipulação de Arquivos de Texto
Escreva um programa Python que leia o arquivo categorias_dados.txt e conte o número de palavras em cada exemplo. Salve os resultados em um novo arquivo chamado contagem_palavras.txt.

```python
conteudo = ""

with open("categorias_dados.txt", "r", encoding="utf-8") as arquivo:
    conteudo = arquivo.read()

#Retirei todos os caracteres indesejados
conteudo = conteudo.replace(':', ',').replace('[', '').replace(']', '').replace('{', '').replace('},', '').replace('}', '').replace('"', '')

#Separei o conteúdo por linha
lista = conteudo.split("\n")

categorias = {
    'Estruturados': [],
    'Semi-Estruturados': [],
    'Desestruturados': [],
    'Temporais': []
}

#Organizei em um dicionário o conteúdo de cada categoria
for item in lista:
    if item !='':
        if item in categorias:
            chave = item
        else:
            categorias[chave].append(item)

categorias['Desestruturados'] = categorias['Desestruturados'][0].split(' ')

#Criei uma função para contar a quantidade de palavras de cada categoria
def contar_palavras(dicionario):
    resposta = {
    'Estruturados': 0,
    'Semi-Estruturados': 0,
    'Desestruturados': 0,
    'Temporais': 0
    }

    for chave in dicionario:
        lista = dicionario[chave]
        lista = ','.join(lista).split(',')
        resposta[chave] = len(lista)

    return resposta

resposta = contar_palavras(categorias)

#Gravei o resultado no TXT
with open("contagem_palavras.txt", "w", encoding="utf-8") as arquivo:
    for chave in resposta:
        arquivo.write(f"{chave}: {resposta[chave]}\n")
```

# Manipulação de Dados Tabulares com CSV
Crie um arquivo CSV chamado dados_alunos.csv com as colunas: nome, idade, curso. Preencha o arquivo com dados fictícios de pelo menos 12 alunos de 3 cursos distintos (Um curso deve ser Ciência de Dados). Escreva um programa Python que leia o arquivo e imprima a média de idade dos alunos.

```python
total_idade = 0
contador = 0

with open('dados_alunos.csv', mode='r', encoding='utf-8') as file:
    next(file)
    for linha in file:
        dados = linha.strip().split(',')
        idade = int(dados[1])
        total_idade += idade
        contador += 1

print(total_idade/contador)
```

# Filtragem de Dados CSV e conversão de CSV para JSON
Usando o arquivo dados_alunos.csv, escreva um programa que filtre e imprima apenas os alunos do curso de Ciência de Dados. Escreva ainda um método que transforme os dados em um arquivo .json e salve como dados_alunos.json

```python
import json
resposta = []

with open('dados_alunos.csv', mode='r', encoding='utf-8') as file:
    next(file)
    for linha in file:
        dados = linha.strip().split(',')
        #Confiro se o curso é ciência de dados
        if dados[2] == 'Ciência de Dados':
            #Se for guardo o resultado em um dicionário
            aluno = {
                'nome': dados[0],
                'idade': int(dados[1]),
                'curso': dados[2]
            }
            resposta.append(aluno)

with open('dados_alunos.json', mode='w', encoding='utf-8') as json_file:
    json.dump(resposta, json_file, indent=4, ensure_ascii=False)
```

# Conversão de JSON para CSV
Escreva um programa Python que leia o arquivo dados_produtos.json e converta os dados para um arquivo CSV chamado produtos.csv.

```python
import json
import csv

with open('dados_produtos.json', 'r') as json_file:
    estoque = json.load(json_file)

with open('produtos.csv', 'w', newline='') as arquivo:
    writer = csv.writer(arquivo)
    writer.writerow(['id', 'nome', 'preco', 'quantidade'])
    
    for produto in estoque:
        writer.writerow([produto['id'], produto['nome'], produto['preco'], produto['quantidade']])
```

# Análise de Dados CSV
Usando o arquivo dados_alunos.csv, escreva um programa que conte o número de alunos em cada curso e salve os resultados em um novo arquivo CSV chamado contagem_cursos.csv.

```python
import csv
resposta = []

dic_qtd_alunos = {}

with open('dados_alunos.csv', mode='r', encoding='utf-8') as file:
    next(file)
    for linha in file:
        col = linha.strip().split(',')
        if col[2] not in dic_qtd_alunos:
            dic_qtd_alunos[col[2]] = 1
        else:
            dic_qtd_alunos[col[2]] += 1


with open("contagem_cursos.csv", "w", newline='', encoding="utf-8") as arquivo:
    writer = csv.writer(arquivo)
    writer.writerow(['curso', 'qtd_alunos'])

    for materia in dic_qtd_alunos:
        writer.writerow([materia, dic_qtd_alunos[materia]])
```
