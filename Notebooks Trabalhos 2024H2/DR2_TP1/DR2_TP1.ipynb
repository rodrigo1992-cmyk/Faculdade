# Manipulação de Arquivos de Texto
Escreva um programa Python que leia o arquivo categorias_dados.txt e conte o número de palavras em cada exemplo. Salve os resultados em um novo arquivo chamado contagem_palavras.txt.

```python
conteudo = ""

with open("categorias_dados.txt", "r", encoding="utf-8") as arquivo:
    conteudo = arquivo.read()

#Retirei todos os caracteres indesejados
conteudo = conteudo.replace(':', ',').replace('[', '').replace(']', '').replace('{', '').replace('},', '').replace('}', '').replace('"', '')

#Separei o conteúdo por linha
lista = conteudo.split("\n")

categorias = {
    'Estruturados': [],
    'Semi-Estruturados': [],
    'Desestruturados': [],
    'Temporais': []
}

#Organizei em um dicionário o conteúdo de cada categoria
for item in lista:
    if item !='':
        if item in categorias:
            chave = item
        else:
            categorias[chave].append(item)

categorias['Desestruturados'] = categorias['Desestruturados'][0].split(' ')

#Criei uma função para contar a quantidade de palavras de cada categoria
def contar_palavras(dicionario):
    resposta = {
    'Estruturados': 0,
    'Semi-Estruturados': 0,
    'Desestruturados': 0,
    'Temporais': 0
    }

    for chave in dicionario:
        lista = dicionario[chave]
        lista = ','.join(lista).split(',')
        resposta[chave] = len(lista)

    return resposta

resposta = contar_palavras(categorias)

#Gravei o resultado no TXT
with open("contagem_palavras.txt", "w", encoding="utf-8") as arquivo:
    for chave in resposta:
        arquivo.write(f"{chave}: {resposta[chave]}\n")
```

# Análise de Dados CSV
Usando o arquivo dados_alunos.csv, escreva um programa que conte o número de alunos em cada curso e salve os resultados em um novo arquivo CSV chamado contagem_cursos.csv.

```python
import csv
resposta = []

dic_qtd_alunos = {}

with open('dados_alunos.csv', mode='r', encoding='utf-8') as arquivo:
    next(arquivo)
    for linha in arquivo:
        col = linha.strip().split(',')
        if col[2] not in dic_qtd_alunos:
            dic_qtd_alunos[col[2]] = 1
        else:
            dic_qtd_alunos[col[2]] += 1


with open("contagem_cursos.csv", "w", newline='', encoding="utf-8") as arquivo:
    writer = csv.writer(arquivo)
    writer.writerow(['curso', 'qtd_alunos'])

    for materia in dic_qtd_alunos:
        writer.writerow([materia, dic_qtd_alunos[materia]])
```
